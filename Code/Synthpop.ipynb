```python
import pandas as pd
import numpy as np
from synthpop import Synthpop
```

```python
df = pd.read_csv("/content/drive/Shareddrives/SI670 Final_Proj/transactions.csv")
```

```python
df.head()
```

```python
df['Date'] = df['TRANSACTION_DATE'].apply(lambda x: x.split('T')[0])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df_2009 = df[df['Date'].dt.year==2009].drop(['OBJECTID',
                                            'TRANSACTION_DATE',
                                            'VENDOR_NAME',
                                            'VENDOR_STATE_PROVINCE',
                                            'MCC_DESCRIPTION'], axis=1)
df_2009.head()
```

```python
df_2009['Date'] = pd.to_numeric(df_2009['Date'])
```

```python
df_2009.head()
```

```python
df_2009.info()
```

```python
spop = Synthpop(visit_sequence=[2, 1, 0])
```

```python
data_types = {"AGENCY": "category",
              "TRANSACTION_AMOUNT": "float",
              "Date": "int"}

spop.fit(df_2009, data_types)
```

```python
spop_syn = spop.generate(len(df_2009))
```

```python
spop_syn.head()
```

```python
spop_syn['Date'] = pd.to_datetime(spop_syn['Date'])
```

```python
spop_syn.head()
```

```python
spop_syn.to_csv('spop_syn.csv', index=False)
```
